```python
#Indeksleme, Dilimleme ve İterasyon
#Numpy dizileri üzerinde indeksleme ve dilimleme işlemleri yapabiliriz.
#Tek boyutlu dizilerde indeksleme ve dilimleme işlemleri Python listeleri ile aynıdır.
a = np.arange(10)**3 #0'dan 9'a kadar olan sayıların küplerini alır.
a
```

```python
a[2]
```

```python
a[2:5]
```

```python
a[:6:2] = 1000 #Dizinin ilk 6 elemanından her ikinci elemanı 1000 yapar.
a
```

```python
a[::-1] #Diziyi tersten yazdırır.
```

```python
for i in a: #Dizinin her bir elemanını yazdırır.
    print(i**(1/3.)) #Küp kökünü alır.
```

```python
#Çok boyutlu dizilerde ise indeksleme ve dilimleme işlemleri biraz daha karmaşıktır. Burada her bir eksen için ayrı ayrı indeksleme yapabiliriz.
def f(x, y):
    return 10 * x + y
b = np.fromfunction(f,(5, 4),dtype=int) #5x4'lük bir dizi oluşturur ve her bir elemanı f fonksiyonu ile belirler.
b
```

```python
b[2, 3] #2. satır ve 3. sütundaki elemanı alır.
```

```python
b[0:5, 1] #Tüm satırlardan 1. sütunu alır.
```

```python
b[:,1] #Tüm satırlardan 1. sütunu alır.
```

```python
b[1:3, :] #1. ve 2. satırları alır.
```

```python
b[-1] #Son satırı alır.
```

```python
# ... operatörü ile eksenleri atlayabiliriz.
#Bu dizide 3 boyutlu bir dizi oluşturalım.
c = np.array([[[0, 1, 2],
               [10, 12, 13]],

              [[100, 101, 102],
               [110, 112, 113]]])
c.shape
```

```python
c[1, ...] #1. bloktaki tüm satır ve sütunları alır. c[1,:,:] ile aynı işlemi yapar.
```

```python
c[..., 2] #Tüm bloklardan 2. sütunu alır. c[:,:,2] ile aynı işlemi yapar.
```
