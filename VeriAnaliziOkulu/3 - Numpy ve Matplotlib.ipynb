## NumPy neden √∂nemli?

NumPy, Python‚Äôda **sayƒ±sal hesap** ve **√ßok boyutlu veri** (vekt√∂r/matris) i≈ülemleri i√ßin temel k√ºt√ºphanedir. √ñzellikle tamamen ya da aƒüƒ±rlƒ±klƒ± olarak **numerik** veriyle √ßalƒ±≈üƒ±rken hƒ±z ve verimlilik saƒülar.

### 1) Hƒ±z ve performans (neden ‚Äúhƒ±zlƒ±‚Äù?)
- NumPy‚Äôƒ±n temel veri yapƒ±sƒ± `ndarray` **tek bir dtype** (homogeneous) ta≈üƒ±r: `int32`, `float64` gibi.
- Tek tip veri sayesinde:
  - Bellekte daha **kompakt** tutulur (daha az ‚Äúek bilgi/metadata‚Äù ta≈üƒ±r).
  - Vekt√∂rize i≈ülemler ve lineer cebir operasyonlarƒ± genelde daha **hƒ±zlƒ±** √ßalƒ±≈üƒ±r.
- Bu y√ºzden bir√ßok k√ºt√ºphane (√∂r. **scikit-learn**) veri giri≈üinde NumPy array‚Äôlerini yaygƒ±n olarak kullanƒ±r; **lineer cebir** ve model matrisi mantƒ±ƒüƒ±na uyumu y√ºksektir.

### 2) NumPy vs Pandas: Ne zaman hangisi?
- **Matematik / model matrisi / b√ºy√ºk sayƒ±sal hesap** ‚Üí **NumPy**
- **Veri temizleme, birle≈ütirme, gruplayƒ±p raporlama** ‚Üí **Pandas**

### 3) Pandas‚Äôta olup NumPy‚Äôda olmayan ‚Äútablo √∂zellikleri‚Äù
Pandas daha ‚ÄúExcel gibi‚Äù √ßalƒ±≈üƒ±r:
- **Kolon isimleri** ve **satƒ±r indeksleri** (√∂rn. `Ya≈ü`, `Cinsiyet` gibi s√ºtunlar; `Ahmet`, `Mehmet` gibi satƒ±r etiketleri)
- **Join/Merge** (tablo birle≈ütirme)
- **GroupBy** (gruplayƒ±p √∂zetleme)
- **Time series resampling** (zaman serisini haftalƒ±k/aylƒ±k toplama vb.)
- **Eksik veri/etiket hizalama** (index √ºzerinden otomatik hizalama)
- **Heterojen tablo**: DataFrame‚Äôde her s√ºtun farklƒ± dtype olabilir

NumPy ise daha ‚Äúham‚Äù ve daha d√º≈ü√ºk seviyelidir:
- Kolon/satƒ±r etiketleri yerine **pozisyon** (0,1,2‚Ä¶) ile √ßalƒ±≈üƒ±r.
- √áok hƒ±zlƒ±dƒ±r ama ‚Äútablo‚Äù ergonomisi Pandas kadar g√º√ßl√º deƒüildir.

### 4) NumPy neden ‚Äúara basamak‚Äù gibi?
Listelerden Pandas‚Äôa giden yolda NumPy kritik bir ge√ßi≈ütir:
- **Vekt√∂r ve matris** mantƒ±ƒüƒ±nƒ± √∂ƒüretir.
- √áok boyutlu veri (`2D`, `3D`, ‚Ä¶) √ºzerinde dilimleme, broadcast, toplama/√ßarpma gibi i≈ülemler i√ßin √ßok i≈ülevseldir.

### 5) G√∂r√ºn√ºm: ‚Äúham‚Äù ve daha teknik
- NumPy dizileri √ßoƒüu zaman i√ß i√ße ge√ßmi≈ü, d√ºzenli (simetrik) listeler gibi g√∂r√ºn√ºr.
- Bu y√ºzden ilk bakƒ±≈üta Pandas‚Äôa g√∂re daha ‚Äúkorkutucu‚Äù gelebilir; ama sayƒ±sal hesap tarafƒ±nda √ßok g√º√ßl√ºd√ºr.

12- Belirli bir dizi (`dizi`) ve istediƒüiniz bir adƒ±m (veya pencere boyutu) i√ßin bir `time_resolved` fonksiyonu yazƒ±n.

Fonksiyon ≈üu √º√ß parametreyi alacak:  
`time_resolved(dizi, pencere_geni≈üliƒüi, kayma)`

√ñrnek olarak,  
`dizi = np.array([1, 2, 6, 4, 5, 4, 2, 7, 8, 9])` i√ßin  
`time_resolved(dizi, 3, 4)` √ßaƒürƒ±ldƒ±ƒüƒ±nda, fonksiyonun √ßƒ±ktƒ±sƒ± `[4, 5, 4]` olmalƒ±dƒ±r.

````python
def time_resolved(dizi, pencere_geni≈üliƒüi, kayma): # Pencere geni≈üliƒüi istenen alt dizinin uzunluƒüu, kayma ise ba≈ülangƒ±√ß indeksi
    alt_dizi=np.array([])
    for i in range(kayma - 1, kayma + pencere_geni≈üliƒüi - 1): # ƒ∞ndeksler kayma-1'den ba≈ülayƒ±p kayma+pencere_geni≈üliƒüi-1'e kadar gider
        alt_dizi=np.append(alt_dizi, dizi[i])
    return alt_dizi
````

### Matplotlib.pyplot ‚Äî Mini Cheat Sheet (tek sayfa)

**Kurulum**
- `import matplotlib.pyplot as plt` ‚Üí pyplot‚Äôu kullanmaya ba≈üla

**√áizim (grafik t√ºr√º se√ß)**
- `plt.plot(x, y)` ‚Üí √ßizgi (trend/zaman serisi)
- `plt.bar(x, h)` ‚Üí bar/s√ºtun (kategori kar≈üƒ±la≈ütƒ±rma)
- `plt.hist(data, bins=...)` ‚Üí histogram (daƒüƒ±lƒ±m)
- `plt.scatter(x, y)` ‚Üí scatter (ili≈üki)

**Anlatƒ± / Okunabilirlik**
- `plt.title("...")` ‚Üí ba≈ülƒ±k
- `plt.xlabel("...")` / `plt.ylabel("...")` ‚Üí eksen isimleri (birim ekle)
- `plt.legend()` ‚Üí seri a√ßƒ±klamasƒ± (√∂nce `label="..."` vermi≈ü ol)
- `plt.grid(True)` ‚Üí ƒ±zgara (opsiyonel)

**D√ºzen**
- `plt.tight_layout()` ‚Üí ta≈ümalarƒ± d√ºzelt, bo≈üluklarƒ± ayarla

**√áƒ±ktƒ±**
- `plt.savefig("fig.png", dpi=300)` ‚Üí dosyaya kaydet (rapor/tez i√ßin)
- `plt.show()` ‚Üí ekranda g√∂ster (genelde en sonda)

**Tipik akƒ±≈ü  
**Draw ‚Üí Label ‚Üí Layout ‚Üí Save ‚Üí Show** ‚úÖ


````python
import matplotlib.pyplot as plt

X = [1,2,3,4]
Y = [10,22,30,40]

plt.plot(X, Y, 'r--1') # 'r--1' kƒ±rmƒ±zƒ± kesikli √ßizgi ile g√∂sterir, Kƒ±rmƒ±zƒ± (red) renk, kesikli √ßizgi (dashed line), 1 i≈üaret√ßi (marker)

plt.xlabel('X label')

plt.ylabel('Y label')

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 200ms aralƒ±kla
t = np.arange(0, 5, 0.2)

plt.plot(t, t, 'r', label='line one', linewidth=2) # 'r' kƒ±rmƒ±zƒ± √ßizgi ile g√∂sterir label ise a√ßƒ±klama i√ßin linewidth ise √ßizgi kalƒ±nlƒ±ƒüƒ±
plt.plot(t, t**2, 'b--^',label='line two', linewidth=2, markersize = 4) # 'b--^' mavi kesikli √ßizgi ile g√∂sterir, mavi (blue) renk, kesikli √ßizgi (dashed line), ^ i≈üaret√ßi (marker)
plt.plot(t,t**3, 'g-o', label='marker', markersize = 4) # 'g-o' ye≈üil √ßizgi ile g√∂sterir, ye≈üil (green) renk, d√ºz √ßizgi (solid line), o i≈üaret√ßi (marker)
plt.legend() # legend fonksiyonu a√ßƒ±klamalarƒ± g√∂sterir
plt.grid(True, alpha=0.5 ,color='k') # grid fonksiyonu ƒ±zgarayƒ± g√∂sterir, alpha saydamlƒ±k, color ise renktir.
plt.show()
````

1- ƒ∞ki veya daha fazla √ßizgiyi, farklƒ± renk ve kalƒ±nlƒ±klarda, a√ßƒ±klamalƒ± (legend) olarak √ßiziniz.

````python
import matplotlib.pyplot as plt

X = [1,2,3,4]
Y = [10,22,30,40]
Z = [3,5,7,9]
Q = [3,15,27,39]

# X ekseni 1 2 3 4 olarak belirlenmi≈ü ve √∂b√ºr diziler bu eksene g√∂re √ßizdirilmi≈ü
plt.plot(X, Y, 'r-o', label="Y", alpha=0.4, linewidth=2) # x = 1 iken y = 10, x = 2 iken y = 22 , x = 3 iken y = 30 , x = 4 iken y = 40
plt.plot(X, Z, 'k--^', label="Z", linewidth=1) # x = 1 iken z = 3, x = 2 iken z = 5 , x = 3 iken z = 7 , x = 4 iken z = 9
plt.plot(X, Q, 'g-+', label="Q", linewidth=1) # x = 1 iken q = 3, x = 2 iken q = 15 , x = 3 iken q = 27 , x = 4 iken q = 39

plt.xlabel('X label',)
plt.ylabel('Y label')
plt.grid(True)

plt.title("Ba≈ülƒ±k")

plt.legend(loc=2) # loc parametresi legend'in konumunu belirler, 2 ise sol √ºst k√∂≈üe, 1 ise saƒü √ºst k√∂≈üe , 3 ise sol alt k√∂≈üe, 4 ise saƒü alt k√∂≈üe

plt.show()
````

2- 0 ile 1000 arasƒ±nda rastgele deƒüerler olu≈üturun ve bu deƒüerleri grafik √ºzerinde g√∂sterin.  
Grafikte, 500‚Äôden k√º√ß√ºk deƒüerler i√ßin mavi, 500‚Äôden b√ºy√ºk deƒüerler i√ßin kƒ±rmƒ±zƒ± renk kullanƒ±n.

````python
import numpy as np

data=np.random.randint(0,1000,50) # 0 ile 1000 arasƒ±nda 50 tane rastgele tam sayƒ± olu≈üturur
x = np.arange(50) # 0'dan 49'a kadar olan sayƒ±lar

# Eƒüer data dizisindeki eleman 500'den b√ºy√ºkse kƒ±rmƒ±zƒ±
data_kƒ±rmƒ±zƒ±=np.array([])
data_kƒ±rmƒ±zƒ±_indeks=np.array([])
# K√º√ß√ºk veya e≈üitse mavi renkte g√∂stermek i√ßin iki ayrƒ± dizi olu≈üturacaƒüƒ±z
data_mavi=np.array([])
data_mavi_indeks=np.array([])

for i in range(len(data)):
    if data[i] > 500: # data dizisindeki eleman 500'den b√ºy√ºkse
        data_kƒ±rmƒ±zƒ± = np.append(data_kƒ±rmƒ±zƒ±, data[i]) # data_kƒ±rmƒ±zƒ± dizisine elemanƒ± ekler
        data_kƒ±rmƒ±zƒ±_indeks = np.append(data_kƒ±rmƒ±zƒ±_indeks, i) # data_kƒ±rmƒ±zƒ±_indeks dizisine elemanƒ±n indeksini ekler
    elif data[i] <= 500: # data dizisindeki eleman 500'den k√º√ß√ºk veya e≈üitse
        data_mavi = np.append(data_mavi, data[i]) # data_mavi dizisine elemanƒ± ekler
        data_mavi_indeks = np.append(data_mavi_indeks, i) # data_mavi_indeks dizisine elemanƒ±n indeksini ekler


plt.plot(x,data,"k--", alpha=0.5) # t√ºm veriyi siyah kesikli √ßizgi ile g√∂sterir
plt.plot(data_mavi_indeks,data_mavi,"b*") # 500'den k√º√ß√ºk verileri mavi yƒ±ldƒ±z ile g√∂sterir
plt.plot(data_kƒ±rmƒ±zƒ±_indeks,data_kƒ±rmƒ±zƒ±,"r*") # 500'den b√ºy√ºk verileri kƒ±rmƒ±zƒ± √º√ßgen ile g√∂sterir
plt.show()
````

3- Numpy alƒ±≈ütƒ±rmalarƒ±nda yazdƒ±ƒüƒ±nƒ±z _time_resolved analysis_ fonksiyonunu, yukarƒ±da yer alan alƒ±≈ütƒ±rmadaki veriye uygulayƒ±n ve sonucu yeni bir grafikte g√∂sterin.

````python
def time_resolved(dizi, pencere_geni≈üliƒüi, kayma):
    alt_dizi = np.array([])
    for i in range(kayma - 1, kayma + pencere_geni≈üliƒüi - 1):
        alt_dizi = np.append(alt_dizi, dizi[i])
    return alt_dizi
````

````python
import matplotlib.pyplot as plt

rand_deƒüerler = np.random.randint(0,1000,50) # 0 ile 1000 arasƒ±nda 50 tane rastgele tam sayƒ± olu≈üturur
print(rand_deƒüerler) # Olu≈üturulan rastgele deƒüerleri yazdƒ±rƒ±r
sayƒ±lar=np.arange(1,51) # 1'den 50'ye kadar olan sayƒ±lar
print(sayƒ±lar) # Olu≈üturulan sayƒ±larƒ± yazdƒ±rƒ±r
plt.plot(sayƒ±lar, rand_deƒüerler, 'r') # Rastgele deƒüerleri kƒ±rmƒ±zƒ± √ßizgi ile g√∂sterir

pencere_geni≈üliƒüi = int(input("Pencere geni≈üliƒüini giriniz.")) # Kullanƒ±cƒ±dan pencere geni≈üliƒüini alƒ±r
kayma = int(input("Ka√ßƒ±ncƒ± indeksten ba≈ülasƒ±n.")) # Kullanƒ±cƒ±dan kayma deƒüerini alƒ±r

alt_dizi = time_resolved(rand_deƒüerler, pencere_geni≈üliƒüi, kayma)
sayƒ±lar = np.arange(kayma, pencere_geni≈üliƒüi + kayma)

print(alt_dizi)
plt.plot(sayƒ±lar,alt_dizi,'g')
plt.show
````
